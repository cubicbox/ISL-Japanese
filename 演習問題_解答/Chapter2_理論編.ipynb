```python
# パッケージのインポート
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0) # 乱数の固定
sns.set_palette('bright')

# データ生成
x = np.arange(0, np.pi , np.pi / 20)
noise = np.random.normal(0, 1, x.size) / 2
y = np.sin(x)
true = pd.DataFrame(data={'x' : x, 'y' : y})
train = pd.DataFrame(data={'x' : x, 'y' : y + noise})

# 描画
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)
#sns.lineplot(x='x', y='y', data=true, errorbar=None, ax=ax)
sns.regplot(x='x', y='y', data=train, order=1, ci=None, ax=ax, label='直線')
sns.regplot(x='x', y='y', data=train, order=2, ci=None, ax=ax, label='2次関数')
sns.regplot(x='x', y='y', data=train, order=10, ci=None, ax=ax, label='10次関数')

# 引用: https://github.com/coxy1989/isl
```
